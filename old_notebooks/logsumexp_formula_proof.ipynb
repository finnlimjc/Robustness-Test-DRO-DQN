```python
import torch
import numpy as np
```

```python
cij = torch.tensor([[[1., 2., 3.],
                     [0., -1., 1.]]])

c = cij.amax(dim=2, keepdim=True)
exponent = torch.exp(cij - c)
inner_exp = exponent.mean(dim=2)
result_else = (torch.log(inner_exp) + c.squeeze(-1)).mean(dim=1)
```

```python
log_meanexp = torch.logsumexp(cij, dim=2, keepdim=True) - np.log(cij.size(2))
result_mod = log_meanexp.mean(dim=1)
```

```python
print("Result (Else Branch):     ", result_else.item())
print("Result (Modified Branch): ", result_mod.item())
print("Difference:               ", abs(result_else - result_mod).item())
```
